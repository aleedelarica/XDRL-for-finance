```python
import pandas as pd
import numpy as np
```

```python
path_result='result1-58.03425266435841.csv'
data = pd.read_csv(path_result,sep=',')
data.drop(['Unnamed: 0'],axis=1,inplace=True)

#Divide Weight column in 6 columns by splitting the string by comma
data[['Weight1','Weight2','Weight3','Weight4','Weight5','Weight6']] = data['Weight'].str.split(',',expand=True)
data.drop(['Weight'],axis=1,inplace=True)

#Divide Price column in 6 columns by splitting the string by comma
data[['Price1','Price2','Price3','Price4','Price5','Price6']] = data['Price'].str.split(',',expand=True)
data.drop(['Price'],axis=1,inplace=True)

data.head()
```

```python
data=data.astype(float)
```

```python

# Save to a new CSV file
data.to_csv('cleaned_'+path_result, index=False)
```
